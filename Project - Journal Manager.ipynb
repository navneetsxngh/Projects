```python
## Project - Journal Manager

while True:
    print("---File Journal Menu---\n" \
    "1. Add Entry\n" \
    "2. Read All Entries\n" \
    "3. Search Entries\n" \
    "4. Exit\n" \
    "Choose Option (1-4)")
    option = int(input("Enter Option: "))

    def write_entry():
        with open("Journal.txt", "a") as file:
            content = input("Enter Entry - ")
            data = file.write(content)
            print(data)
    def read_entry():
        with open("Journal.txt", "r") as file:
            data = file.read()
            print(data)

    def find_entry():
        keyword = input("Search - ")
        with open("Journal.txt", "r") as file:
            all_lines = file.readlines()
            for line in all_lines:
                if keyword.lower() in line.lower():
                    print(line.strip())
                    return
        print("No Matching Entry")

    if option == 1:
        write_entry()
    elif option == 2:
        read_entry()
    elif option == 3:
        find_entry()
    elif option == 4:
        break
    else:
        print("Invalid Option")
```
