```python
import os
import pandas as pd
```

```python
BASE_DIR = os.getcwd()
PROJECT_ROOT = os.path.dirname(BASE_DIR)
DATA_PATH = os.path.join(PROJECT_ROOT, "data", "ultimate_student_productivity_dataset_5000.csv")
```

```python
df = pd.read_csv(DATA_PATH)
df.head()
```

```python
df.info()
```

```python
df['gender'].value_counts()
```

```python
df['academic_level'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
cat_cols = df.select_dtypes(include='str').columns

for col in cat_cols:
    df[col] = lb.fit_transform(df[col])

df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
SAVE_PATH = os.path.join(PROJECT_ROOT, "data", 'preprocessed.csv')
df.to_csv(SAVE_PATH, index=False)
```
